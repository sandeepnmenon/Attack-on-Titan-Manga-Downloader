```python
from concurrent.futures import ThreadPoolExecutor
from bs4 import BeautifulSoup
import requests,os
```

```python
def download_chapter(num):
    print(f"Thread - {num} started")
    html_page = requests.get(f"https://readshingekinokyojin.com/manga/shingeki-no-kyojin-chapter-{num}/").content
    soup = BeautifulSoup(html_page)
    images = []
    folder = f"./chapter-{num}/"
    if not os.path.exists(os.path.dirname(folder)):
        os.makedirs(os.path.dirname(folder))
        print(f"{folder} created..")
    cnt = 1
    for img in soup.findAll('img'):
        image_url = img.get('src')
        response = requests.get(image_url)
        open(f"{folder}/{cnt}.png", "wb").write(response.content)
        print(f"{folder}/{cnt}.png created")
        cnt+=1
    print(f"Thread - {num} finished")
```

## Set start and end chapter range to download

```python
start_chapter = 1
end_chapter = 132
```

```python
with ThreadPoolExecutor(max_workers=10) as executor:
    for i in range(start_chapter,end_chapter + 1):
        executor.submit(download_chapter,i)
```
